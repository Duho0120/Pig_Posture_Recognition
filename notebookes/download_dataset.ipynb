# Kaggle 데이터셋 다운로드

이 노트북은 캐글 API의 **가장 최신 인증 방식(API Token)**을 사용하여 `multi-view-pig-posture-recognition` 대회 데이터셋을 다운로드합니다.

**중요:** [캐글 대회 페이지](https://www.kaggle.com/competitions/multi-view-pig-posture-recognition/data)에서 **Rules(동의)**가 완료되어야 정상적으로 다운로드됩니다.

```python
import os

# 발급받으신 'KGAT_...'로 시작하는 토큰을 아래 따옴표 안에 붙여넣으세요.
os.environ['KAGGLE_API_TOKEN'] = 'KGAT_c70ab721a9ee257dd38c9d7d1e6bc6c7'

print("Kaggle API 토큰이 설정되었습니다.")
```

```python
from kaggle.api.kaggle_api_extended import KaggleApi

try:
    # API 인증 및 인스턴스 생성
    api = KaggleApi()
    api.authenticate()
    print("Kaggle API 인증 성공!")
    
    # 대회 이름 설정
    competition_name = 'multi-view-pig-posture-recognition'
    
    # 대회 접근 가능 여부 미리 확인
    print(f"'{competition_name}' 대회 접근 확인 중...")
    api.competition_list_files(competition_name)
    print("대회 접근 권한 확인 완료!")
    
except Exception as e:
    print(f"\n인증 또는 권한 오류가 발생했습니다.\n{e}")
    print("\n[해결 방법]")
    print("1. 토큰이 정확히 복사되었는지 확인해 주세요.")
    print("2. 대회 페이지(https://bit.ly/40YshT7)에서 'Rules' 동의를 하셨는지 확인해 주세요.")
```

```python
download_path = '../data'
os.makedirs(download_path, exist_ok=True)

print(f"{competition_name} 데이터셋 다운로드 중...")

# 대회 데이터셋 다운로드
try:
    api.competition_download_files(competition_name, path=download_path)
    print(f"다운로드 완료! 파일 경로: {download_path}")
except Exception as e:
    print(f"다운로드 중 오류 발생: {e}")
```

```python
import zipfile
import os

zip_file_path = os.path.join(download_path, f'{competition_name}.zip')
extract_path = os.path.join(download_path, competition_name)

os.makedirs(extract_path, exist_ok=True)

if os.path.exists(zip_file_path):
    print("압축 해제 중...")
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print(f"압축 해제 완료! 저장 위치: {os.path.abspath(extract_path)}")
else:
    print(f"오류: {zip_file_path} 파일을 찾을 수 없습니다.")
```
